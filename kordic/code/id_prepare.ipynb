#### 네이버 아이디 표 준비 

오픈프로에서 다운받아 naver_id, type, ohw, shoulder 이렇게 네개로 있는 액셀 파일 가져와서 idiom, proverb, ohw 별로 나놔서 작업 

```python
import pandas as pd

id_path = "/Users/anjonghyeon/Desktop/KU/민족문화연구원 사전실/openpro/kordic/asset/naver_id_1212.xlsx"
target = pd.read_excel(id_path, engine='openpyxl')
target
```

```python
target = target.dropna(subset=['naver_id'])
target['shoulder'] = target['shoulder'].astype(int)
target
```

```python
target_idiom = target[target['type'] == '숙어']
target_proverb = target[target['type'] == '속담']
target_ohw = target[target['type'] == '단어']

target_idiom = target_idiom.sort_values(by=['ohw', 'shoulder'])
target_proverb = target_proverb.sort_values(by=['ohw', 'shoulder'])
target_ohw = target_ohw.sort_values(by=['ohw', 'shoulder'])
```

```python
target_idiom.rename(columns={'ohw':'idiom'}, inplace=True)
target_proverb.rename(columns={'ohw':'proverb'}, inplace=True)
```

```python
target_idiom.to_csv('../asset/idiom_id.csv', index=False)
target_proverb.to_csv('../asset/proverb_id.csv', index=False)
target_ohw.to_csv('../asset/ohw_id.csv', index=False)
```
