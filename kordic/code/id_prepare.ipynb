#### 네이버 아이디 표 준비 

오픈프로에서 다운받아 naver_id, type, ohw, shoulder 이렇게 네개로 있는 액셀 파일 가져와서 idiom, proverb, ohw 별로 나놔서 작업 

```python
import pandas as pd

# id_path를 바꿔주세요.
id_path = "../asset/naver_id.xlsx"
target = pd.read_excel(id_path, engine='openpyxl')
target
```

```python
target = target.dropna(subset=['엔트리ID'])
target['어깨번호'] = target['어깨번호'].astype(int)
target
```

```python
target_idiom = target[target['표제어종류'] == '숙어']
target_proverb = target[target['표제어종류'] == '속담']
target_ohw = target[target['표제어종류'] == '단어']

target_idiom = target_idiom.sort_values(by=['제공처표제어', '어깨번호'])
target_proverb = target_proverb.sort_values(by=['제공처표제어', '어깨번호'])
target_ohw = target_ohw.sort_values(by=['제공처표제어', '어깨번호'])
```

```python
target_idiom.rename(columns={'제공처표제어':'idiom'}, inplace=True)
target_proverb.rename(columns={'제공처표제어':'proverb'}, inplace=True)
```

```python
target_idiom.to_csv('../asset/idiom_id.csv', index=False)
target_proverb.to_csv('../asset/proverb_id.csv', index=False)
target_ohw.to_csv('../asset/ohw_id.csv', index=False)
```
