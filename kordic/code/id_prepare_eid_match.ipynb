```python
KORDIC = pd.merge(
    KORDIC,
    NAVER[["제공처표제어", "어깨번호", "엔트리ID"]], 
    on=["제공처표제어", "어깨번호"],
    how="left",
    suffixes=("", "_NAVER")
)

KORDIC["엔트리ID"] = KORDIC["엔트리ID_NAVER"].combine_first(KORDIC["엔트리ID"])
KORDIC.drop(columns=["엔트리ID_NAVER"], inplace=True)
```
