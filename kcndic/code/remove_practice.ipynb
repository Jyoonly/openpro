#### 오픈프로 형식에 변형한 한한중 사전 데이터에서 중복값을 제거할 때 사용할 코드!!

```python
import pandas as pd

file1 = "/Users/anjonghyeon/Desktop/KU/DIC_OPENPRO/kcndic/resource/found.xlsx"
target = pd.read_excel(file1, engine='openpyxl')
target
```

우선 지금 쓸데없는 칼럼들이 많아 몇개 지워주도록 하겠다..<br><br>
표제어를 나타내는 칼럼은 headword만 남기고 headword2, headword3은 제거한다.<br><br>
중국어 뜻풀이 부분도 sense만 남기고 sense2는 지운다.<br><br>
연관어도 relword만 남기고 relword2는 지운다. <br><br>
일단 용례를 나타내는 exam칼럼도 exam만 남기고 exam2, exam3, exam4는 지우도록 하겠다..

```python
target = target.drop(columns=['headword2', 'headword3', 'sense2', 'exam2', 'exam3', 'exam4', 'relword2'])
target = target.sort_values(by=['headword', 'pos', 'senseNum'])
target = target.reset_index(drop=True)
target
```

##### "\N" 값들 처리
본격적으로 중복값 제거에 앞서, 현재 신규어라 용례칸은 모두 비어있어 "\N"로 들어가 있다.<br><br>
그래서 이 값들을 모두 빈칸으로 만들어주어야한다. <br><br>
네이버 오픈프로에 올릴때 그냥 진짜로 "\N"로 인식하더라.. 

```python
target['exam'] = target['exam'].str.replace(r"\N", "", regex=False)
target['reltype'] = target['reltype'].str.replace(r"\N", "", regex=False)
target['relword'] = target['relword'].str.replace(r"\N", "", regex=False)
target = target.fillna("")
target
```

##### 중복값 처리 
자 이제 본격적으로 중복값 처리를 할건데 <br><br>
우선 headword 칼럼이 기준이 된다. headword 칼럼의 값이 돌면서 중복된 값이 있으면 빈값으로 바꿔줘야한다. 

```python
target.loc[target.duplicated(subset=['headword']), 'headword'] = ""
target
```

그리고 이제 origin도 중복값을 없애줘야 하는데.. <br><br>
origin의 경우 headword가 비어있으면 그냥 그 행의 값도 비워주면 된다. 

```python
condition = target['headword'] == ""
target.loc[condition, ['origin']] = ""
target
```

자.. 이제 거의 다 해간다... <br><br>
근데.. useform 칼럼은 그냥 죄다 없네..  나중에 추가사항 생기면 추가하도록 하겠다.. <br><br>
이제 품사(pos), 뜻번호(senseNum), 문형(moonhung), 뜻풀이(sense), 한국어 뜻풀이(senseKR) 다섯개 남았다. <br><br>
이 다섯개는 서로 연관되어 있어 같이 없애도록 하겠다. <br><br>
우선 품사가 전 행의 품사와 같으면서 현재 행의 headword가 빈값이면 지워야 한다..

```python
before_pos = target['pos'][0]
for i in range(1, len(target)):
    if (target['headword'][i] == "" and target['pos'][i] == before_pos):
        target['pos'][i] = ""
    else: 
        before_pos = target['pos'][i]
target.iloc[20:30, :]
```

이제 senseNum을 없애보자 <br><br>
senseNum은 만약 품사가 없는데 전번호와 같으면 없애면 된다.

```python
before_num = 1
for i in range(len(target)):
    if (target['pos'][i] == ""):
        if (target['senseNum'][i] == before_num):
            target['senseNum'][i] = ""
            target['moonhung'][i] = ""
            target['sense'][i] = ""
            target['senseKR'][i] = ""
        else:
            before_num = target['senseNum'][i]
    else:
        before_num = 1
target.head(20)
```

마지막으로 해야할 부분은.. 연관어 부분이다.. <br><br>
연관어 부분을 볼 때는 만약 senseNum 부분이 없다면 중복되는 행일 것이다. 이 때 relword와 reltype의 값이 같다면 없애면 된다. 

```python
reltype_list = []
relword_list = []
for i in range(len(target)):
    if (target['senseNum'][i] == ""):
        if target['reltype'][i] in reltype_list:
            if target['relword'][i] in relword_list:
                target['reltype'][i] = ""
                target['relword'][i] = ""
            else: 
                relword_list.append(target['relword'][i])
        else: 
            reltype_list.append(target['reltype'][i])
            if target['relword'][i] not in relword_list:
                relword_list.append(target['relword'][i])
    else:
        reltype_list = []
        relword_list = []
        reltype_list.append(target['reltype'][i])
        relword_list.append(target['relword'][i])
target.head(20)
```

이제 수정된 target 데이터프레임을 내보내주면 되겠다!!

```python
result_path = "/Users/anjonghyeon/Desktop/KU/DIC_OPENPRO/kcndic/result/final.xlsx"
target.to_excel(result_path, index=False)
print("finish..")
```
